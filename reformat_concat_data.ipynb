```python
def reformat(df: pd.DataFrame, dataset_type: str):
    df = df.drop(columns=['Unnamed: 0'], axis=1)
    df['label'] = df['index_spans'].apply(lambda x: 1 if x != '[]' else 0)
    df.drop(columns=['index_spans'], inplace=True)
    df.rename(columns={'content': 'text'}, inplace=True)
    df['dataset_type'] = dataset_type
    return df 
    
    
```

```python
vihos_train = reformat(vihos_train, 'train')
vihos_dev = reformat(vihos_dev, 'dev')
vihos_test = reformat(vihos_test, 'test')
```

# **Phản động dataset**

```python
pd_train = pd.read_csv('data\phan-dong\\train_data.csv')
pd_dev = pd.read_csv('data\phan-dong\\valid_data.csv')
pd_test = pd.read_csv('data\phan-dong\\test_data.csv')
```

```python
def phan_dong_reformat(df: pd.DataFrame, dataset_type: str):
    df.drop(axis= 1, columns= ['Unnamed: 0', 'Unnamed: 0.1', 'url', 'checked', 'content_normalize', 'length', 'combine_text', 'text', 'text_split'], inplace= True)
    df.rename(columns= {'content': 'text'}, inplace= True)
    df['dataset_type'] = dataset_type
    return df
```

```python
pd_train = phan_dong_reformat(pd_train, 'train')
pd_dev = phan_dong_reformat(pd_dev, 'dev')
pd_test = phan_dong_reformat(pd_test, 'test')
```

```python
def chunk_into_4_sentence(text: str):
    # print(text)
    sentences = text.split('.')
    res = []
    for i in range(0, len(sentences), 3):
        res.append('.'.join(sentences[i: i + 3]))
        if len(res) == 10:
            break 
    return res

tmp = chunk_into_4_sentence(pd_train.loc[0, 'text'])
print(tmp)
print(len(tmp))

    
        
```

```python

def create_chunk_df(df: pd.DataFrame):
    text_col = []
    label_col = [] 
    dataset_type_col = []
    for index, row in df.iterrows():
        label = row['label']  
        chunks = chunk_into_4_sentence(row['text'])
        dataset_type = row['dataset_type']
        for chunk in chunks:
            if chunk == ' ' or chunk == '': continue
            text_col.append(chunk)
            label_col.append(label)
            dataset_type_col.append(dataset_type)
        # print(index)
    return pd.DataFrame({'text': text_col, 'label': label_col, 'dataset_type': dataset_type_col})
        
```

```python
pd_train = create_chunk_df(pd_train)
pd_dev = create_chunk_df(pd_dev)
pd_test = create_chunk_df(pd_test)
```
